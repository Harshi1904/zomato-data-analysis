```python
# Keep only values that look like ratings (e.g., 4.1, 3.8)

df['rate'] = df['rate'].astype(str)

df['rate'] = df['rate'].str.extract(r'(\d+\.\d+)')

df['rate'] = pd.to_numeric(df['rate'], errors='coerce')

print(df['rate'].dtype)
```

```python
df['votes'] = pd.to_numeric(df['votes'], errors='coerce')
```

```python
df[['rate','votes','approx_cost(for two people)']].dtypes
```

```python
df = df.dropna(subset=['rate'])
```

```python
print(df.shape)
df[['name','rate']].head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_df = df[['rate','votes','approx_cost(for two people)']]

plt.figure(figsize=(8,5))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
from wordcloud import WordCloud

text = " ".join(
    df['cuisines']
    .dropna()
    .astype(str)
)

wordcloud = WordCloud(
    width=1000,
    height=500,
    background_color='white'
).generate(text)

plt.figure(figsize=(12,6))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.scatterplot(
    x='approx_cost(for two people)',
    y='rate',
    data=df
)

plt.title("Price vs Rating")
plt.show()
```

# Key Findings

1. Most restaurants are concentrated in a few popular locations.
2. North Indian and Chinese cuisines are among the most common cuisines.
3. Online ordering is available for many restaurants.
4. Most ratings fall between 3.5 and 4.5.
5. Cost and rating have a weak correlation.

# Recommendations

1. Focus marketing efforts on popular restaurant locations.
2. Promote highly rated restaurants with fewer votes.
3. Improve cuisine-based recommendations.
4. Encourage online ordering adoption.
5. Use ratings and votes for restaurant ranking.
